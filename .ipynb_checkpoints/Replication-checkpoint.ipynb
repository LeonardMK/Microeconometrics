### Figure 4: Daily Wage Corruption Measures with Discontinuous Polynomial Fits

```python
# Create aggregate analysis data set
fig_4_dw = usedata.loc[usedata["anydw"] == 1, ["day", "daysdwoff", "daysdw"]].groupby("day").sum()
fig_4_pr = usedata.loc[usedata["anypr"] == 1, ["day", "rateproff", "ratepr"]].groupby("day").sum()
fig_4_all = pd.concat([fig_4_dw, fig_4_pr], axis = 1)
fig_4_all.reset_index(inplace = True)
fig_4_all["diff_dw"] = fig_4_all["daysdwoff"] - fig_4_all["daysdw"]
fig_4_all["diff_pr"] = fig_4_all["rateproff"] - fig_4_all["ratepr"]
fig_4_all["shock"] = pd.Series(fig_4_all["day"] > 120)
fig_4_all["shock_day"] = fig_4_all["shock"]*(fig_4_all["day"] - 121)
fig_4_all["day2"] = fig_4_all["day"]**2
fig_4_all["shock_day2"] = fig_4_all["shock"]*(fig_4_all["day"] - 121)**2
```

```python
# Regression models
dw_1 = smf.ols("diff_dw ~ day + shock + shock_day", fig_4_all)
dw_2 = smf.ols("diff_dw ~ day + day2 + shock + shock_day + shock_day2", fig_4_all)
pr_1 = smf.ols("diff_pr ~ day + shock + shock_day", fig_4_all)
pr_2 = smf.ols("diff_pr ~ day + day2 + shock + shock_day + shock_day2", fig_4_all)

# Fitting models
dw1_fit = dw_1.fit()
dw2_fit = dw_2.fit()
pr1_fit = pr_1.fit()
pr2_fit = pr_2.fit()

# Getting predictions plus SE
dw_pred = dw1_fit.get_prediction(fig_4_all[["day", "shock", "shock_day"]]).summary_frame(alpha = 0.05)
dw_pred_2 = dw2_fit.get_prediction(fig_4_all[["day", "day2", "shock", "shock_day", "shock_day2"]]).summary_frame(alpha = 0.05)
pr_pred = pr1_fit.get_prediction(fig_4_all[["day", "shock", "shock_day"]]).summary_frame(alpha = 0.05)
pr_pred_2 = pr2_fit.get_prediction(fig_4_all[["day", "day2", "shock", "shock_day", "shock_day2"]]).summary_frame(alpha = 0.05)
```

### Table 2: Wage Shock Effects on Project Composition

```python
# Function that calculates covariance matrix in the dimensions upid, day and intersection
def cov_2way(model, data = usedata, clusters = ["upid", "day"]):
    
    # Define interaction term
    data[clusters[0] + "_" + clusters[1]] = data[clusters[0]].astype("str") + data[clusters[1]].astype("str")
    data[clusters[0] + "_" + clusters[1]] = data[clusters[0] + "_" + clusters[1]].astype("category")
    
    # Fit all regression models
    crude = model.fit()
    upid = model.fit(cov_type = "cluster", cov_kwds = {"groups": data.loc[model.data.row_labels, clusters[0]]})
    day = model.fit(cov_type = "cluster", cov_kwds = {"groups": data.loc[model.data.row_labels, clusters[1]]})
    upid_day = model.fit(cov_type = "cluster", cov_kwds = {"groups": data.loc[model.data.row_labels, (clusters[0] + "_" + clusters[1])]})
    
    # Adjust the original covariance matrix
    crude.cov_params = upid.cov_params() + day.cov_params() - upid_day.cov_params()
    crude.HC0_se = crude.cov_params
    crude.cov_HC0 = np.array(crude.cov_params)
        
    # Return adjust matrix
    return crude
```

```python
def results_lm(models, coef = None, r = 4, r_R2 = 3, R2 = True, TE = "Day", FE = "District", Obs = True, TW = None, labels = None):
    
    # Creating header
    col_labels = list(range(1, len(models) + 1))
    header = pd.DataFrame(columns = col_labels)
    
    # Getting coefficients for every coef
    if coef is None:
        coef = models[len(models) - 1].params
        
    # Test list
    re = []
    
    # Matrix to save coefficients
    # Need indices first
    y = list(np.repeat(coef, 2))
    for i in range(1, 2 * len(coef), 2):
        y[i] = y[i] + "SE"
    
    M = pd.DataFrame(np.empty((2 * len(coef), len(models)), "str"), columns = col_labels, index = y)
    
    for i in range(len(coef)):
        re.append(coef[i] + " = 0")
    
    for i in range(len(models)):
        
        # Check if all coefs are actually in the model
        index = []
        
        for j in coef:
            index.append(j in models[i].params)
        
        test = (models[i].t_test(", ".join(np.array(re)[index].tolist()), cov_p = models[i].cov_params))
        
        # Create a df
        res_df = pd.DataFrame(np.array([test.effect, test.sd, test.pvalue]), index = ["Coef", "SD", "P"], columns = np.array(coef)[index])
        res_df = res_df.transpose()
        
        # Create a star column
        res_df["star"] = pd.cut(res_df["P"], bins = [0, 0.01, 0.05, 0.1, 1], labels = ["***", "**", "*", ""])
        
        # Rounding coefficients and SD
        res_df[["Coef", "SD"]] = res_df[["Coef", "SD"]].round(r)
                
        # Insert into M-data-frame
        for j in res_df.index:
            
            M.loc[j, i + 1] = res_df.loc[j, "Coef"].astype("str") + " " + res_df.loc[j, "star"]
            M.loc[j + "SE", i + 1] = "(" + res_df.loc[j, "SD"].astype("str") + ")"
            
    # Change indices of M
    M.reset_index(inplace = True)
    for i in range(1, len(M["index"]), 2):
        M.loc[i, "index"] = " "
        
    # Replace index if labels is given
    M.loc[list(range(0, len(M["index"]), 2)), "index"] = labels
        
    # Set index again
    M.set_index("index", inplace = True)
        
    # Append to header
    header = header.append(M, ignore_index = False)
        
    # R2
    if R2:
        R2_se = pd.DataFrame([i.rsquared for i in models], columns = ["R2"], index = list(range(1, len(models) + 1))).transpose().round(r_R2)
        header = header.append(R2_se, ignore_index = False)

    # TE
    if TE is not None:
        TE_se = pd.DataFrame(TE, columns = ["Time controls"], index = list(range(1, len(models) + 1))).transpose()
        header = header.append(TE_se, ignore_index = False)
        
    # FE
    if FE is not None:
        FE_se = pd.DataFrame(FE, columns = ["Fixed Effects"], index = list(range(1, len(models) + 1))).transpose()
        header = header.append(FE_se, ignore_index = False)
        
    # Obs
    if Obs:
        Obs_se = pd.DataFrame([i.nobs for i in models], columns = ["Observations"], index = list(range(1, len(models) + 1))).astype("int").transpose()
        header = header.append(Obs_se, ignore_index = False)
        
    if TW is not None:
        TW_se = pd.DataFrame(TW, columns = ["Time window (months)"], index = list(range(1, len(models) + 1))).transpose()
        header = header.append(TW_se, ignore_index = False)
        
    header.rename_axis("Regressors", inplace = True)
        
    return header
```

```python
# First regression model
fdwfrac1 = smf.ols("fdwfrac ~ shock + day + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", usedata)

# Add did
fdwfrac2 = smf.ols("fdwfrac ~ shock + day + did + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", usedata)

# Add day2
fdwfrac3 = smf.ols("fdwfrac ~ shock + day + day2 + did + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", usedata)

# Fit models using the formula for Cameron and Miller 2015.
m1 = cov_2way(fdwfrac1)
m2 = cov_2way(fdwfrac2)
m3 = cov_2way(fdwfrac3)
m4 = cov_2way(fdwfrac3, clusters = ["ubid", "day"])
```

```python
Table2 = results_lm([m1, m2, m3, m4], ["shock", "day", "day2"], r = 3, r_R2 = 3, R2 = True, TE = None, FE = ["No", "District", "District", "District"], Obs = True, TW = None, labels = ["Shock", "Day", "Day squared"])
Table2.to_html("tables/Table2.html", justify = "center")
```

```python
# Shock
t3_A_1 = smf.ols("daysdwoff ~ daysdw + shock + day + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3",
                 usedata.loc[usedata["anydw"] == 1])
t3_A_1_fit = cov_2way(t3_A_1)

# Shock + did
t3_A_2 = smf.ols("daysdwoff ~ daysdw + shock + day + C(did) + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", 
                 usedata.loc[usedata["anydw"] == 1])
t3_A_2_fit = cov_2way(t3_A_2)

# Shock, RD controls
t3_A_3 = smf.ols("daysdwoff ~ daysdw + shock + day + shock_day + C(did) + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", 
                 usedata.loc[usedata["anydw"] == 1])
t3_A_3_fit = cov_2way(t3_A_3)

# Calculating intercept difference at the time the shock occurs for all RD-type regs
t3_A_3_fit.params["shock"] = t3_A_3_fit.params["shock"] + t3_A_3_fit.params["shock_day"]*121
t3_A_3_fit.cov_params.loc["shock", "shock"] = (t3_A_3_fit.cov_params.loc["shock", "shock"] + 121*121*t3_A_3_fit.cov_params.loc["shock_day", "shock_day"]
                                               + 2*121*t3_A_3_fit.cov_params.loc["shock", "shock_day"])

# DiD with alwaysdw
t3_A_4 = smf.ols("daysdwoff ~ daysdw + shock + shock_alwaysdw + alwaysdw +  day + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3",
                 usedata.loc[usedata["anydw"] == 1])
t3_A_4_fit = cov_2way(t3_A_4)

# DiD with alwaysdw + did
t3_A_5 = smf.ols("daysdwoff ~ daysdw + shock + shock_alwaysdw + alwaysdw + day + C(did) + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3",
                 usedata.loc[usedata["anydw"] == 1])
t3_A_5_fit = cov_2way(t3_A_5)

# DiD with alwaysdw + did + RD
t3_A_6 = smf.ols("daysdwoff ~ daysdw + shock + shock_alwaysdw + alwaysdw + day + shock_day + C(did) + day + holiday + res_gender + res_sbc + res_tribe + season1 + season2 + dayofmonth + dayofmonth2 + dayofmonth3", 
                 usedata.loc[usedata["anydw"] == 1])
t3_A_6_fit = cov_2way(t3_A_6)

# Calculating intercept difference at the time the shock occurs for all RD-type regs
t3_A_6_fit.params["shock"] = t3_A_6_fit.params["shock"] + t3_A_6_fit.params["shock_day"]*121
t3_A_6_fit.cov_params.loc["shock", "shock"] = (t3_A_6_fit.cov_params.loc["shock", "shock"] + 121*121*t3_A_6_fit.cov_params.loc["shock_day", "shock_day"]
                                               + 2*121*t3_A_6_fit.cov_params.loc["shock", "shock_day"])
```

```python
# Creating the table

# Put results into a list
res_list = [orbdo_mle1, orbdo_mle2, orcol_mle1, orcol_mle2]

# Create pd.DataFrame
res = pd.DataFrame(np.empty((12, 4)), dtype = "str")

for i in range(4):
    
    model = res_list[i]
    
    # Get significance stars of model    
    stars = pd.cut(pd.Series(sp.stats.norm.cdf(np.abs(model.x/np.sqrt(model.hess_inv.diagonal())))), bins = [0, 0.95, 0.975, 0.995, 1], labels = ["", "*", "**", "***"])
    
    # Iterate of coefficients
    for j in range(len(model.x)):
        
        # Insert into res
        res.iloc[2 * j, i] = model.x[j].round(3).astype("str") + stars[j]
        res.iloc[1 + 2 * j, i] = np.sqrt(model.hess_inv[j, j]).round(3).astype("str")
        
# Change columns and index labels
res.rename_axis("Regressor", inplace = True)
res.rename(index = {0: "Shock", 1: "", 2: "Koraput", 3: "", 4: "Gajapati", 5: "", 6: "Rayagada", 7: "", 8: "Day", 9: "", 10: "Day squared", 11: ""},
          columns = {0: "BDO", 1: "BDO", 2: "Collector", 3: "Collector"}, inplace = True)

# Replace values in (12, 1) and (12, 3) with ""
res.iloc[[10, 11], [0, 2]] = ""

# Save as latex table
res.to_html("tables/Table8.html", justify = "center")
```

```python
# Estimating GG bottom line effects & SE
dw_coef = t3_B_2_fit.params["shock"]
dw_coef_fdwall = t3_B_2_fit.params["shock_fdw_all"]
dw_coef_fdwsome = t3_B_2_fit.params["shock_fdw_some"]

# Variance
dw_var = t3_B_2_fit.cov_params.loc["shock", "shock"]
dw_fdwallvar = t3_B_2_fit.cov_params.loc["shock_fdw_all", "shock_fdw_all"]
dw_fdwsomevar = t3_B_2_fit.cov_params.loc["shock_fdw_some", "shock_fdw_some"]
dw_shock_fdwall = t3_B_2_fit.cov_params.loc["shock", "shock_fdw_all"]
dw_shock_fdwsome = t3_B_2_fit.cov_params.loc["shock", "shock_fdw_some"]
dw_fdwall_fdwsome = t3_B_2_fit.cov_params.loc["shock_fdw_some", "shock_fdw_all"]
dw_n = usedata.loc[(usedata["anydw"] == 1) & (usedata["shock"] == 1) & (usedata["fdwfrac"].isna() == False)].shape[0]
fdwfrac_sumall = usedata.loc[(usedata["dwactive"] == 1) & (usedata["shock"] == 1) & (usedata["fdwfrac"] == 1), "fdwfrac"].sum()
fdwfrac_sumsome = usedata.loc[(usedata["dwactive"] == 1) & (usedata["shock"] == 1) & (usedata["fdwfrac"] < 1) & (usedata["fdwfrac"] > 0), "fdwfrac"].sum()
dw_meanwage = 66.8

# Coeffs from pr regression
pr_n = usedata.loc[(usedata["anypr"] == 1) & (usedata["shock"] == 1)].shape[0]

# Calculations
act_dw = (dw_coef * dw_n + dw_coef + dw_coef_fdwall * fdwfrac_sumall + dw_coef_fdwsome * fdwfrac_sumsome) * dw_meanwage

drop_dw = (dw_coef_fdwall * fdwfrac_sumall + dw_coef_fdwsome * fdwfrac_sumsome) * dw_meanwage

# Coef from pr
pr_coef = t4_A_2_fit.params["shock"]
drop_pr = pr_coef * pr_n

# Calculate effect
print(np.round((-drop_dw - drop_pr)/(act_dw - drop_dw - drop_pr), 2))

# Variance calculations
act_dw_var = ((dw_n ** 2) * dw_var + (fdwfrac_sumall ** 2) * dw_fdwallvar + (fdwfrac_sumsome ** 2) * dw_fdwsomevar + 2 * dw_n * fdwfrac_sumall * dw_shock_fdwsome + 
             2 * fdwfrac_sumsome * fdwfrac_sumall * dw_fdwall_fdwsome) * (dw_meanwage ** 2)

drop_dw_var = ((fdwfrac_sumall ** 2) * dw_fdwallvar + (fdwfrac_sumsome ** 2) * dw_fdwsomevar + 2 + fdwfrac_sumsome *fdwfrac_sumall * dw_fdwall_fdwsome) * (dw_meanwage) ** 2

pr_var = t4_A_2_fit.cov_params.loc["shock", "shock"]
drop_pr_var = pr_var * (pr_n ** 2)
num_var = drop_dw_var + drop_pr_var
denom_var = act_dw_var + drop_dw_var + drop_pr_var

full_var = ((drop_dw + drop_pr) ** 2)/((act_dw - drop_dw - drop_pr) **2) * (num_var / ((drop_dw + drop_pr) ** 2) + denom_var / ((act_dw - drop_dw - drop_pr) ** 2))

np.round((full_var ** 0.5) * 1.6, 2)
```
